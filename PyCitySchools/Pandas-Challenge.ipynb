```python
import pandas as pd
from pathlib import Path
```

```python
# load CSV files
schools_df = pd.read_csv("schools_complete.csv")
students_df = pd.read_csv("students_complete.csv")
# Merge datasets
school_data_complete = pd.merge(students_df, schools_df, how="left", on="school_name")
```

```python
# Total number of unique schools
total_schools = schools_df["school_name"].nunique()
print("Total number of schools:", total_schools)
```

```python
# Total number of students
total_students = students_df["Student ID"].count()
print("Total number of students:", total_students)
```

```python
# Total budget
total_budget = schools_df["budget"].sum()
print(f"Total Budget: ${total_budget:,.0f}")
```

```python
# Average math and reading scores
average_math_score = students_df["math_score"].mean()
average_reading_score = students_df["reading_score"].mean()
print("Total Average Math Score:", average_math_score)
print("Total Average Reading Score:", average_reading_score)
```

```python
# % Passing math and reading
passing_math = students_df[students_df["math_score"] >= 70]["Student ID"].count()
passing_reading = students_df[students_df["reading_score"] >= 70]["Student ID"].count()
overall_passing = students_df[(students_df["math_score"] >= 70) & (students_df["reading_score"] >= 70)]["Student ID"].count()
```

```python
percent_passing_math = (passing_math / total_students) * 100
percent_passing_reading = (passing_reading / total_students) * 100
percent_overall_passing = (overall_passing / total_students) * 100
```

```python
# Create District Summary DataFrame
district_summary = pd.DataFrame({
    "Total Schools": [total_schools],
    "Total Students": [total_students],
    "Total Budget": [f"${total_budget:,.0f}"],  # Format the budget as a string with commas and dollar sign
    "Average Math Score": [average_math_score],
    "Average Reading Score": [average_reading_score],
    "% Passing Math": [percent_passing_math],
    "% Passing Reading": [percent_passing_reading],
    "% Overall Passing": [percent_overall_passing],
})

district_summary
```

## SCORES BY SCHOOL SPENDING

```python
import pandas as pd

# Load the CSV files
schools_df = pd.read_csv("schools_complete.csv")
students_df = pd.read_csv("students_complete.csv")

# Merge the two datasets on school name
school_data_complete = pd.merge(students_df, schools_df, how="left", on="school_name")

# Calculate total students, budget, and scores for each school
per_school_counts = schools_df.set_index("school_name")["size"]
per_school_budget = schools_df.set_index("school_name")["budget"]
per_student_budget = per_school_budget / per_school_counts

# Group the merged dataset by school name for average scores
per_school_math = school_data_complete.groupby("school_name")["math_score"].mean()
per_school_reading = school_data_complete.groupby("school_name")["reading_score"].mean()

# Calculate passing rates
passing_math = school_data_complete[school_data_complete["math_score"] >= 70]
passing_reading = school_data_complete[school_data_complete["reading_score"] >= 70]
overall_passing = school_data_complete[
    (school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)
]

per_school_passing_math = passing_math.groupby("school_name")["Student ID"].count() / per_school_counts * 100
per_school_passing_reading = passing_reading.groupby("school_name")["Student ID"].count() / per_school_counts * 100
overall_passing_rate = overall_passing.groupby("school_name")["Student ID"].count() / per_school_counts * 100

# Add school type
school_types = schools_df.set_index("school_name")["type"]

# Create a school summary DataFrame
per_school_summary = pd.DataFrame({
    "School Type": school_types,
    "Total Students": per_school_counts,
    "Total School Budget": per_school_budget,
    "Per Student Budget": per_student_budget,
    "Average Math Score": per_school_math,
    "Average Reading Score": per_school_reading,
    "% Passing Math": per_school_passing_math,
    "% Passing Reading": per_school_passing_reading,
    "Overall Passing Rate": overall_passing_rate
})

# Add spending ranges (Per Student)
spending_bins = [0, 585, 630, 645, 680]
spending_labels = ["<$585", "$585-630", "$630-645", "$645-680"]
per_school_summary["Spending Ranges (Per Student)"] = pd.cut(
    per_school_summary["Per Student Budget"],
    bins=spending_bins,
    labels=spending_labels
)

# Formatting the budget columns
per_school_summary["Total School Budget"] = per_school_summary["Total School Budget"].map("${:,.2f}".format)
per_school_summary["Per Student Budget"] = per_school_summary["Per Student Budget"].map("${:,.2f}".format)

# Create a copy of the school summary for later aggregations
school_spending_df = per_school_summary.copy()

# Display all 15 schools in the dataset
pd.set_option("display.max_rows", 15)  # Ensure all rows are displayed
print(school_spending_df)
```

## Calculate averages for desired columns

```python
import pandas as pd
```

```python
## Import datasets
schools_df = pd.read_csv("schools_complete.csv")
students_df = pd.read_csv("students_complete.csv")
```

```python
## Merge datasets
school_data_complete = pd.merge(students_df, schools_df, how="left", on="school_name")
```

```python
per_school_counts = schools_df.set_index("school_name")["size"]
per_school_budget = schools_df.set_index("school_name")["budget"]
per_student_budget = per_school_budget / per_school_counts
school_types = schools_df.set_index("school_name")["type"]
```

```python
per_school_math = school_data_complete.groupby("school_name")["math_score"].mean()
per_school_reading = school_data_complete.groupby("school_name")["reading_score"].mean()
```

```python
## Calculate passing rates
passing_math = school_data_complete[school_data_complete["math_score"] >= 70]
passing_reading = school_data_complete[school_data_complete["reading_score"] >= 70]
overall_passing = school_data_complete[
    (school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)
]
per_school_passing_math = passing_math.groupby("school_name")["Student ID"].count() / per_school_counts * 100
per_school_passing_reading = passing_reading.groupby("school_name")["Student ID"].count() / per_school_counts * 100
overall_passing_rate = overall_passing.groupby("school_name")["Student ID"].count() / per_school_counts * 100
```

```python
per_school_summary = pd.DataFrame({
    "School Type": school_types,
    "Total Students": per_school_counts,
    "Total School Budget": per_school_budget,
    "Per Student Budget": per_student_budget,
    "Average Math Score": per_school_math,
    "Average Reading Score": per_school_reading,
    "% Passing Math": per_school_passing_math,
    "% Passing Reading": per_school_passing_reading,
    "% Overall Passing": overall_passing_rate
})
```

```python
## Spending ranges (per student) 
spending_bins = [0, 585, 630, 645, 680]
spending_labels = ["<$585", "$585-630", "$630-645", "$645-680"]
per_school_summary["Spending Ranges (Per Student)"] = pd.cut(
    per_school_summary["Per Student Budget"], 
    bins=spending_bins, 
    labels=spending_labels
)
```

```python
school_spending_df = per_school_summary.copy()
```

```python
spending_math_scores = school_spending_df.groupby(
    "Spending Ranges (Per Student)", observed=False
)["Average Math Score"].mean()

spending_reading_scores = school_spending_df.groupby(
    "Spending Ranges (Per Student)", observed=False
)["Average Reading Score"].mean()

spending_passing_math = school_spending_df.groupby(
    "Spending Ranges (Per Student)", observed=False
)["% Passing Math"].mean()

spending_passing_reading = school_spending_df.groupby(
    "Spending Ranges (Per Student)", observed=False
)["% Passing Reading"].mean()

overall_passing_spending = school_spending_df.groupby(
    "Spending Ranges (Per Student)", observed=False
)["% Overall Passing"].mean()
```

```python
spending_summary = pd.DataFrame({
    "Average Math Score": spending_math_scores,
    "Average Reading Score": spending_reading_scores,
    "% Passing Math": spending_passing_math,
    "% Passing Reading": spending_passing_reading,
    "% Overall Passing": overall_passing_spending
})
```
